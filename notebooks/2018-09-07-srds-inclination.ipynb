```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Config matplotlib
font = {'size'   : 22}

matplotlib.rc('font', **font)
```

```python
# Prefix simulation
prefix_simulation = "MVS_Vulcan-J-"

# Raw data
data_path = "/home/sandro/Projetos/MVS_Vulcan-J/"

# Number of clones
n_clones = 96
```

```python
# Define variables
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune", "Vulcan"]
```

```python
def plot_oe( plt_name, y_inf = 0, y_sup = 1.0):
    plt.figure(figsize=(16,5))

    os.chdir(data_path)
    for clone in range(1,n_clones + 1):
        os.chdir(prefix_simulation + str(clone))
        data = np.loadtxt(plt_name + ".txt")
        plt.plot(data[:,0], data[:,3]) # Inclination
        os.chdir("..")
    os.chdir(data_path)

    plt.ylim([y_inf,y_sup])
    plt.xlabel("Time [a]")
    plt.ylabel("Inclination")

    plt.show()
```

## Comparing Earth and Venus

```python
plt.figure(figsize=(16,10))

os.chdir(data_path)

plt.subplot(211)
plt.ylabel("Eccentrity")
for clone in range(1,n_clones + 1):
    os.chdir(prefix_simulation + str(clone))
    venus = np.loadtxt("Venus" + ".txt")
    plt.plot(venus[:,0], venus[:,2])
    os.chdir("..")

plt.subplot(212)
plt.ylabel("Eccentrity")
for clone in range(1,n_clones + 1):
    os.chdir(prefix_simulation + str(clone))
    earth = np.loadtxt("Earth" + ".txt")
    plt.plot(earth[:,0], earth[:,2])
    os.chdir("..")
    
os.chdir(data_path)

#plt.ylim([0, 0.1])
plt.xlabel("Time [a]")


plt.show()
```

## Selecting the simulation of higher eccentricity.

```python
# Discovering the simulation that generated the greatest eccentricity in Venus.
os.chdir(data_path)
ecc_max = 0
for clone in range(1,n_clones + 1):
    os.chdir(prefix_simulation + str(clone))
    venus = np.loadtxt("Venus" + ".txt")
    ecc = venus[:,2].max()
    if ecc > ecc_max:
        ecc_max = ecc
        sim = prefix_simulation + str(clone)
    os.chdir("..")

os.chdir(data_path)

print("Max eccentricity: {}".format(ecc_max))
print("Simulation: {}".format(sim))
```

```python
# Discovering the simulation that generated the greatest eccentricity in Earth.
os.chdir(data_path)
ecc_max = 0
for clone in range(1,n_clones + 1):
    os.chdir(prefix_simulation + str(clone))
    earth = np.loadtxt("Earth" + ".txt")
    ecc = earth[:,2].max()
    if ecc > ecc_max:
        ecc_max = ecc
        sim = prefix_simulation + str(clone)
    os.chdir("..")

os.chdir(data_path)

print("Max eccentricity: {}".format(ecc_max))
print("Simulation: {}".format(sim))
```
