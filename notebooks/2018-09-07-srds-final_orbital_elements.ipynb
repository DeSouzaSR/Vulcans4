```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Config matplotlib
font = {'size'   : 22}

matplotlib.rc('font', **font)
```

```python
# Prefix simulation
prefix_simulation = "MVS_Vulcan-J-"

# Raw data
data_path = "/home/sandro/Projetos/MVS_Vulcan-J/"

# Number of clones
n_clones = 96
```

```python
# Define variables
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune", "Vulcan"]
```

## Eccentricity

```python
inner = planets[0:4]
inner.append(planets[8])
```

```python
plt.figure(figsize=(8,8))

os.chdir(data_path)
for clone in range(1,n_clones + 1):
    os.chdir(prefix_simulation + str(clone))
    for planet in inner[0:5]:
        data = np.loadtxt(planet + ".txt")
        plt.plot(data[0,1], np.mean(data[:,2]), 'k.')
    os.chdir("..")
os.chdir(data_path)

plt.plot(0.38, 0.2056, 'r^') # Mercury
plt.plot(0.72, 0.0068, 'r^') # Venus
plt.plot(1.00, 0.0167, 'r^') # Earth
plt.plot(1.52, 0.0937, 'r^') # Mars

plt.xlabel("Semimajor axis [au]")
plt.ylabel("Eccentricity")

plt.show()
```

## Inclination

```python
plt.figure(figsize=(8,8))

os.chdir(data_path)
for clone in range(1,n_clones + 1):
    os.chdir(prefix_simulation + str(clone))
    for planet in inner[0:5]:
        data = np.loadtxt(planet + ".txt")
        plt.plot(data[0,1], np.mean(data[:,3]), 'k.')
    os.chdir("..")
os.chdir(data_path)

plt.plot(0.38, 7, 'r^') # Mercury
plt.plot(0.72, 3, 'r^') # Venus
plt.plot(1.00, 0, 'r^') # Earth
plt.plot(1.52, 2, 'r^') # Mars

plt.xlabel("Semimajor axis [au]")
plt.ylabel("Inclination")
```
